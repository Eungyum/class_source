```python
import pandas as pd
import numpy as np
import bar_chart_race as bar
```

```python
df = pd.read_csv('C:/k_digital/Data Analysis with Python/data/premierLeague_tables_1992-2017.csv')
df
```

```python
df2 = df[['season', 'team', 'points']]
df2.head()
```

```python
# 행 'season', 열 'team', 데이터 'points'
raw = df2.pivot_table(values='points', index='season', columns='team')
raw.head()
```

```python
# NaN값을 0으로 대체
raw = raw.fillna(0)
raw.head()
```

```python
# 시즌별 점수가 누적해서 더해지도록 수정
raw.iloc[:] = raw.iloc[:].cumsum()
```

```python
raw.head()
```

```python
bar.bar_chart_race(df=raw,
                  n_bars=10,
                  figsize = (6, 4),
                  sort='desc',
                  title='premierLeague_tables_1992-2017')
```

```python
import bar_chart_race as bcr
df = bcr.load_dataset('covid19_tutorial')
bcr.bar_chart_race(
    df=df,
    filename='covid19_horiz.mp4',
    orientation='h',
    sort='desc',
    n_bars=6,
    fixed_order=False,
    fixed_max=True,
    steps_per_period=10,
    interpolate_period=False,
    label_bars=True,
    bar_size=.95,
    period_label={'x': .99, 'y': .25, 'ha': 'right', 'va': 'center'},
    period_fmt='%B %d, %Y',
    period_summary_func=lambda v, r: {'x': .99, 'y': .18,
                                      's': f'Total deaths: {v.nlargest(6).sum():,.0f}',
                                      'ha': 'right', 'size': 8, 'family': 'Courier New'},
    perpendicular_bar_func='median',
    period_length=500,
    figsize=(5, 3),
    dpi=144,
    cmap='dark12',
    title='COVID-19 Deaths by Country',
    title_size='',
    bar_label_size=7,
    tick_label_size=7,
    shared_fontdict={'family' : 'Helvetica', 'color' : '.1'},
    scale='linear',
    writer=None,
    fig=None,
    bar_kwargs={'alpha': .7},
    filter_column_colors=False)  
```
